```python
"""
cargaremos un dataset para explorarlo yanalizarlo
"""
#importaciones de librerias
import pandas as pd
import numpy as np

#creacion de dataframe
df_1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CAPITULO VII/Axie Infinity.csv')
df_2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CAPITULO VII/Celo.csv')
df_3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CAPITULO VII/Dash.csv')
```

```python
"""
revisaremos si hay valores nulos
"""
lista_de_df = [df_1, df_2, df_3]

for _ in lista_de_df:
  print("dataframe")
  print(_.isnull().sum())
  print("")
```

```python
"""
analizaremos las subidas de cada cripto por separado
"""
#solo nos quedamos con las columnas que nos interesan en otro dataframe
df_modifi_1 = df_1[['Date', 'Open', 'Close']]
df_modifi_2 = df_2[['Date', 'Open', 'Close']]
df_modifi_3 = df_3[['Date', 'Open', 'Close']]

print(df_modifi_1)
```

```python
"""
agregaremos una columna con la variacion neta del precio
"""
"""
variacion neta del precio es cierre - apertura
"""
df_modifi_1['Variacion'] = df_modifi_1['Close'] - df_modifi_1['Open']
df_modifi_2['Variacion'] = df_modifi_2['Close'] - df_modifi_2['Open']
df_modifi_3['Variacion'] = df_modifi_3['Close'] - df_modifi_3['Open']

print(df_modifi_1)
```

```python
"""
graficaremos cada uno de los dataframes
"""
import matplotlib.pyplot as plt

#grafico 1
df_modifi_1.plot(x = 'Date', y = 'Variacion', title='Variacion neta')
plt.show()

#grafico 2
df_modifi_2.plot(x = 'Date', y = 'Variacion', title='Variacion neta')
plt.show()

#grafico 3
df_modifi_3.plot(x = 'Date', y = 'Variacion', title='Variacion neta')
plt.show()
```

```python
"""
esto no se entiende mucho asi que podemos trabajar con el promedio semanal
"""
row_ini = 0
row_fin = 30

#creacion de dataframe vacio
columnas = ['Date', 'Variacion']
df_prom_1 = pd.DataFrame([[np.nan, np.nan]], columns = columnas)

#limite final
limite = ((df_modifi_1.shape[0] // 30) * 30)

while df_prom_1.loc[df_prom_1.shape[0] - 1, 'Date'] != df_modifi_1.loc[df_modifi_1.shape[0] - 1, 'Date']:

  promedio_variacion = df_modifi_1['Variacion'].iloc[row_ini : row_fin].mean()
  fecha = df_modifi_1.loc[row_fin, 'Date']

  df_aux = pd.DataFrame([[fecha, promedio_variacion]], columns = columnas)
  df_prom_1 = pd.concat([df_prom_1, df_aux], ignore_index=True)

  row_ini = row_fin + 1
  if (row_ini >= limite):
    row_fin = df_modifi_1.shape[0] - 1
  else:
    row_fin = row_fin + 30

df_prom_1 = df_prom_1.drop(0)

print(df_prom_1)

ax = df_prom_1.plot(x = 'Date', y = 'Variacion', title='Variacion neta')
ax.axhline(y = df_prom_1['Variacion'].mean(), color = 'red', linestyle='--',)

df_prom_1['SMA_3m'] = df_prom_1['Variacion'].rolling(window=3).mean()
prediccion = df_prom_1['SMA_3m'].iloc[-1]

df_prom_1['SMA_3m'].plot(ax = ax, kind='line', color='orange', linestyle='--')

plt.show()

print(df_prom_1)
```
