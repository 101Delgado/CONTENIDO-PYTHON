# ORDENAMIENTO BURBUJA üéà

```python
"""
es un algoritmo de ordenamiento simple que funciona revisando repetidamente la lista a ordenar
y comparando cada par de elementos adyacentes,
intercambi√°ndolos de posici√≥n si est√°n en el orden incorrecto.
"""

# Como funciona?

"""
1) Se compara el primer elemento con el segundo. Si el primero es mayor que el segundo, se intercambian.
2) Se compara el segundo con el tercero. Si el segundo es mayor, se intercambian.
3) Este proceso contin√∫a hasta el final de la lista. Al finalizar la primera pasada, el elemento m√°s grande estar√° en la √∫ltima posici√≥n.
4) La siguiente pasada hace lo mismo, pero se detiene una posici√≥n antes, ya que el √∫ltimo elemento ya est√° en su lugar correcto.
5) El proceso se repite, reduciendo el rango de comparaci√≥n en cada pasada, hasta que no se realiza ning√∫n intercambio.
"""
```

```python
# Existe un problema...

#que pasa cuando las listas son muy grandes

"""
Su complejidad de tiempo en el peor y caso promedio es de O(n2), donde n es el n√∫mero de elementos.
"""

# Veremos un ejemplo de su codigo
```

```python
def ordenamiento_burbuja(lista):
    """
    Ordena una lista de n√∫meros de forma ascendente utilizando el algoritmo de ordenamiento burbuja.
    """
    n = len(lista)

    # Recorre todos los elementos de la lista
    for i in range(n):
        # La variable 'intercambio_hecho' optimiza el algoritmo.
        # Si no se hace ning√∫n intercambio en una pasada, la lista ya est√° ordenada.
        intercambio_hecho = False

        # √öltimos i elementos ya est√°n en su lugar, as√≠ que no es necesario volver a compararlos.
        for j in range(0, n - i - 1):

            # Compara elementos adyacentes
            if lista[j] > lista[j + 1]:

                # Intercambia los elementos si est√°n en el orden incorrecto
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                print(f"ordenando {lista}")
                intercambio_hecho = True

        # Si no se hizo ning√∫n intercambio en esta pasada, la lista est√° ordenada
        if not intercambio_hecho:
            break

    return lista

# Ejemplo de uso:
mi_lista = [64, 34, 25, 12, 22, 11, 90, 50]
lista_ordenada = ordenamiento_burbuja(mi_lista)
print("Lista ordenada:", lista_ordenada)
```

```python
"""
hecho de forma recursiva
"""
def burbuja_recursiva(lista, n):
    """
    Ordena una lista de forma ascendente utilizando el algoritmo de ordenamiento burbuja de manera recursiva.

    Args:
        lista (list): La lista de elementos a ordenar.
        n (int, opcional): El n√∫mero de elementos a considerar en la pasada actual.
                            Por defecto, se toma la longitud completa de la lista.
    """

    print(f"n vale: {n}")


    # Caso base de la recursi√≥n: si la lista tiene 1 elemento o menos, ya est√° ordenada.
    if n <= 1:
        return

    # Realiza una pasada del ordenamiento burbuja
    for i in range(n - 1):
        if lista[i] > lista[i + 1]:
            # Intercambia los elementos si est√°n en el orden incorrecto
            lista[i], lista[i + 1] = lista[i + 1], lista[i]

    # Llama a la funci√≥n recursivamente para ordenar el resto de la lista.
    # El elemento m√°s grande de la pasada anterior ya est√° en su lugar,
    # por lo que el siguiente llamado se enfoca en 'n-1' elementos.
    burbuja_recursiva(lista, n - 1)

# ---

# Ejemplo de uso
mi_lista = [64, 34, 25, 12, 22, 11, 90,50]
print("Lista original:", mi_lista)

burbuja_recursiva(mi_lista, len(mi_lista))

print("Lista ordenada:", mi_lista)
```

# ALGORITMO DE SELECCION

```python
def ordenamiento_por_seleccion(lista):
    """
    Ordena una lista de n√∫meros de forma ascendente utilizando el algoritmo de ordenamiento por selecci√≥n.
    """
    n = len(lista)

    # Recorre cada elemento de la lista para colocarlo en su posici√≥n correcta
    for i in range(n):
        # Asume que el elemento actual es el m√≠nimo
        indice_minimo = i

        # Busca el elemento m√≠nimo en el resto de la lista (sublista desordenada)
        for j in range(i + 1, n):
            if lista[j] < lista[indice_minimo]:
                indice_minimo = j

        # Intercambia el elemento m√≠nimo encontrado con el primer elemento de la sublista desordenada
        # Este paso coloca el elemento m√°s peque√±o en su posici√≥n ordenada
        print(lista)
        lista[i], lista[indice_minimo] = lista[indice_minimo], lista[i]

    return lista

# Ejemplo de uso:
mi_lista = [64, 25, 12, 22, 11]
lista_ordenada = ordenamiento_por_seleccion(mi_lista)
print("Lista ordenada:", lista_ordenada)
```

RECURSIVAMENTE... (NO SE SI ESO ES UNA PALABRA üôÇ)

```python
def seleccion_recursiva(lista, inicio=0):
    """
    Ordena una lista de forma ascendente utilizando el algoritmo de ordenamiento por selecci√≥n
    de manera recursiva.

    Args:
        lista (list): La lista de elementos a ordenar.
        inicio (int): El √≠ndice de la sublista desde donde buscar el elemento m√≠nimo.
    """
    # Caso base: si el √≠ndice de inicio es igual a la longitud de la lista,
    # significa que ya hemos recorrido todos los elementos y la lista est√° ordenada.
    if inicio >= len(lista) - 1:
        return

    # Encuentra el √≠ndice del elemento m√≠nimo en la sublista desde 'inicio'
    indice_minimo = inicio
    for i in range(inicio + 1, len(lista)):
        if lista[i] < lista[indice_minimo]:
            indice_minimo = i
    print(lista)
    # Intercambia el elemento m√≠nimo con el elemento en la posici√≥n 'inicio'
    lista[inicio], lista[indice_minimo] = lista[indice_minimo], lista[inicio]

    # Llama a la funci√≥n recursivamente para ordenar el resto de la lista.
    # El nuevo 'inicio' es 'inicio + 1'.
    seleccion_recursiva(lista, inicio + 1)
    print(f"hola soy {inicio}")
# ---

# Ejemplo de uso
mi_lista = [64, 25, 12, 22, 11, 4]
print("Lista original:", mi_lista)

seleccion_recursiva(mi_lista)

print("Lista ordenada:", mi_lista)
```

# ALGORITMO DE INSERCCION

```python
"""
es un algoritmo de ordenamiento simple y eficiente para peque√±as cantidades de datos. A diferencia de otros m√©todos,
este algoritmo construye la lista final ordenada un elemento a la vez,
tomando los elementos de la lista de entrada y "insert√°ndolos" en la posici√≥n correcta de la lista ya ordenada.
"""
```

```python
def ordenamiento_por_insercion(lista):
    """
    Ordena una lista de n√∫meros de forma ascendente utilizando el algoritmo de ordenamiento por inserci√≥n.
    """
    # Recorre la lista desde el segundo elemento hasta el final
    for i in range(1, len(lista)):

        # 'clave' es el elemento actual que queremos insertar en la sublista ordenada
        clave = lista[i]

        # 'j' es el √≠ndice del √∫ltimo elemento de la sublista ordenada
        j = i - 1

        # Mueve los elementos de la sublista ordenada que son mayores que la 'clave'
        # un lugar a la derecha para hacer espacio.
        while j >= 0 and clave < lista[j]:
            lista[j + 1] = lista[j]
            j -= 1
        print(lista)
        # Inserta la 'clave' en su posici√≥n correcta
        lista[j + 1] = clave
        print(lista)
    return lista

# Ejemplo de uso:
mi_lista = [12, 11, 13, 5, 6]
lista_ordenada = ordenamiento_por_insercion(mi_lista)
print("Lista ordenada:", lista_ordenada)
```

```python
def insercion_recursiva(lista, n):
    """
    Ordena una lista de forma ascendente utilizando el algoritmo de ordenamiento por inserci√≥n
    de manera recursiva.

    Args:
        lista (list): La lista de elementos a ordenar.
        n (int): El n√∫mero de elementos a considerar en la llamada actual.
    """
    # Caso base de la recursi√≥n: si la sublista tiene un solo elemento o menos,
    # ya est√° ordenada y no hay nada que hacer.
    if n <= 1:
        return

    # Llama a la funci√≥n recursivamente para ordenar los primeros 'n-1' elementos.
    insercion_recursiva(lista, n - 1)
    print(f"hola soy {n}")
    # El 'n-1' elemento (√∫ltimo de la sublista desordenada) es la clave a insertar.
    clave = lista[n - 1]
    j = n - 2

    # Mueve los elementos de la sublista ordenada que son mayores que la 'clave'
    # un lugar a la derecha.
    while j >= 0 and clave < lista[j]:
        lista[j + 1] = lista[j]
        j -= 1

    print(lista)
    # Inserta la clave en su posici√≥n correcta.
    lista[j + 1] = clave

# ---

# Ejemplo de uso
mi_lista = [12, 11, 13, 5, 6, 8]
n = len(mi_lista)
print("Lista original:", mi_lista)

insercion_recursiva(mi_lista, n)

print("Lista ordenada:", mi_lista)
```

```python
def merge_sort(lista):
    """
    Ordena una lista de forma ascendente utilizando el algoritmo de ordenamiento por fusi√≥n.
    """
    # Caso base de la recursi√≥n: si la lista tiene un solo elemento o est√° vac√≠a,
    # ya est√° ordenada.
    if len(lista) > 1:

        # 1. DIVIDIR: Encuentra el punto medio de la lista y la divide en dos mitades.
        mitad = len(lista) // 2
        mitad_izquierda = lista[:mitad]
        mitad_derecha = lista[mitad:]

        # Llama a merge_sort de forma recursiva en cada mitad
        merge_sort(mitad_izquierda)
        merge_sort(mitad_derecha)

        # 2. FUSIONAR: Combina las dos mitades ordenadas en una sola lista
        i = j = k = 0

        print(f"antes del ciclo while {lista} con mitad izquierda {mitad_izquierda} con mitad derecha {mitad_derecha}")
        # Fusiona las sublistas temporales en la lista original
        while i < len(mitad_izquierda) and j < len(mitad_derecha):
            print(f"se compara {mitad_izquierda[i]} con {mitad_derecha[j]}")
            if mitad_izquierda[i] < mitad_derecha[j]:
                lista[k] = mitad_izquierda[i]
                i += 1
            else:
                lista[k] = mitad_derecha[j]
                j += 1
            k += 1
        print(f"despues del primer while {lista} con mitad izquierda {mitad_izquierda} con mitad derecha {mitad_derecha}")
        # Carga los elementos restantes de la mitad izquierda, si los hay
        while i < len(mitad_izquierda):
            lista[k] = mitad_izquierda[i]
            i += 1
            k += 1
        print(f"despues del segundo while {lista} con mitad izquierda {mitad_izquierda} con mitad derecha {mitad_derecha}")
        # Carga los elementos restantes de la mitad derecha, si los hay
        while j < len(mitad_derecha):
            lista[k] = mitad_derecha[j]
            j += 1
            k += 1
        print(f"despues del tercer while {lista} con mitad izquierda {mitad_izquierda} con mitad derecha {mitad_derecha}\n")
# ---

# Ejemplo de uso:
mi_lista = [38,13, 27,13, 43, 3, 9, 82, 10]
print("Lista original:", mi_lista)

merge_sort(mi_lista)

print("Lista ordenada:", mi_lista)
```

# EL ORDENAMIENTO RAPIDO

```python
"""
es un algoritmo de ordenamiento muy popular y eficiente que tambi√©n se basa en la t√©cnica de "divide y vencer√°s".
Fue desarrollado por Tony Hoare y, en la pr√°ctica,
suele ser m√°s r√°pido que otros algoritmos de ordenamiento de su misma complejidad, como el ordenamiento por fusi√≥n.
"""

"""
La complejidad de tiempo de Quicksort en el caso promedio y mejor caso es de O(nlogn)
"""
```

```python
def ordenamiento_rapido(lista):
    """
    Ordena una lista de forma ascendente utilizando el algoritmo de ordenamiento r√°pido.
    """
    # Llama a la funci√≥n auxiliar que realiza la recursi√≥n
    _ordenamiento_rapido_aux(lista, 0, len(lista) - 1)

def _ordenamiento_rapido_aux(lista, inicio, fin):
    """
    Funci√≥n auxiliar recursiva para el ordenamiento r√°pido.
    """
    if inicio < fin:
        # 1. PARTIR: Se obtiene el √≠ndice del pivote despu√©s de la partici√≥n.
        indice_particion = particion(lista, inicio, fin)

        # 2. VENCER: Se hacen llamadas recursivas a las dos sublistas
        _ordenamiento_rapido_aux(lista, inicio, indice_particion - 1)  # Sublista izquierda
        _ordenamiento_rapido_aux(lista, indice_particion + 1, fin)     # Sublista derecha

def particion(lista, inicio, fin):
    """
    Funci√≥n de partici√≥n que reorganiza la lista alrededor de un pivote.
    En este caso, se elige el √∫ltimo elemento como pivote.
    """
    # 1. SELECCIONAR EL PIVOTE: En este caso, el √∫ltimo elemento
    pivote = lista[fin]

    # 'i' es el √≠ndice del elemento m√°s peque√±o, que se usar√° para el intercambio
    i = inicio - 1


    # Recorre la lista desde el inicio hasta el fin - 1
    for j in range(inicio, fin):
        # Si el elemento actual es menor o igual al pivote
        if lista[j] <= pivote:
            # Incrementa el √≠ndice del elemento m√°s peque√±o
            i += 1
            # Intercambia los elementos
            lista[i], lista[j] = lista[j], lista[i]
            print(lista)


    # Intercambia el pivote con el elemento en la posici√≥n correcta (i + 1)
    lista[i + 1], lista[fin] = lista[fin], lista[i + 1]
    print(f"con el cambio de pivote {lista}")
    # Devuelve la posici√≥n final del pivote
    return i + 1

# ---

# Ejemplo de uso:
mi_lista = [10, 7, 8, 2, 1, 5] #-> [2,7,8,10,1,5] -> [2,1,8,10,7,5] -> [2,1,5,10,7,8] ->[1,2,5,10,7,8] -> [1,2,5,7,10,8]
# -> [1,2,5,7,8,10]
print("Lista original:", mi_lista)

ordenamiento_rapido(mi_lista)

print("Lista ordenada:", mi_lista)
```
