```python
"""
pregunta 1
"""
"""
una funcion:
es un bloque de codigo reutilizable que podemo llamar a lo largo de nuestro codigo
principal
"""

# primera forma
def suma_impar(n):
  acumulador = 0

  for _ in range(n+1):
    if (_ % 2) != 0:
      print(f"el {_} es impar")
      acumulador += _

  print(f"el resultado es {acumulador}")
  return acumulador


suma = suma_impar(10)
print(f"este es el valor de la variable {suma}")

# segunda forma
def suma_impar_recursiva(n):

    if n % 2 != 0 :
      if n < 0:
        print(f"no se procesan numero negativos")
        return 0
      else:
        print(f"{n} es impar")
        return n + suma_impar_recursiva(n - 1)
    elif n == 0:
      return 0
    else:
      return suma_impar_recursiva(n - 1)
```

```python
"""
pregunta 2
"""
def Codificar(oracion, n):
  oracion = oracion.upper()
  oracion_nueva = ""

  for i in oracion:
    if ord(i) >= 65 and ord(i) <= 90:
      numero = ord(i) + n
      if numero > 90:
        numero -= 26
      oracion_nueva += chr(numero)
    else:
      oracion_nueva += i

  return oracion_nueva

def Decodificar(oracion, n):
  oracion = oracion.upper()
  oracion_nueva = ""

  for i in oracion:
    if ord(i) >= 65 and ord(i) <= 90:
      numero = ord(i) - n
      if numero < 65:
        numero += 26
      oracion_nueva += chr(numero)
    else:
      oracion_nueva += i

  return oracion_nueva

str_1 = Codificar("el perroz", 5)
print(f"oracion codficiada {str_1}")
print(f"oracion decodificada {Decodificar(str_1,5)}")
```

guardar diccionarios en python


```python
import json
import os

"""
les aconsejo que corran esto en vs code o en anaconda
aqui arroja un error debido a que no corre directamente en el entorno virtual de la pc
"""

def guardar_diccionario_en_json_en_escritorio(nombre_archivo, diccionario_a_guardar):
    """
    Guarda un diccionario de Python en un archivo JSON en el Escritorio del usuario.

    Args:
        nombre_archivo (str): El nombre del archivo (ej. 'mis_datos.json').
        diccionario_a_guardar (dict): El diccionario que se desea guardar.
    """
    try:
        #coloquen su ruta de carpeta
        ruta_usuario = 'C:/Users/usuario/carpeta/'

        ruta_completa = ruta_usuario + nombre_archivo

        # 4. Abrir y guardar el archivo.
        with open(ruta_completa, 'w', encoding='utf-8') as archivo_json:
            # json.dump() escribe el diccionario en el archivo, con indentación para legibilidad.
            json.dump(diccionario_a_guardar, archivo_json, indent=4, ensure_ascii=False)

        print(f"✅ ¡Éxito! Diccionario guardado en:")
        print(f"👉 {ruta_completa}")

    except Exception as e:
        print(f"❌ Ocurrió un error inesperado al guardar el archivo: {e}")

# --- Uso del código ---

# 1. Define el nombre del archivo.
nombre_del_archivo = 'mis_datos_del_escritorio.json'

# 2. Crea tu diccionario de ejemplo.
mis_datos = {
    "proyecto": "Guardar en JSON",
    "version": 1.0,
    "destino": "Escritorio del Usuario",
    "elementos": ["Ruta Absoluta", "Módulo OS", "JSON"],
    "fecha": "2025-09-06"
}

# 3. Llama a la función para guardar el diccionario en el Escritorio.
guardar_diccionario_en_json_en_escritorio(nombre_del_archivo, mis_datos)
```
