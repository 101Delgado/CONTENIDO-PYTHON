```python
# primera forma de hacer la funcion

def sumar_elementos_1(n):
  a = 0

  if n >= 0:

    for _ in range(n+1):
      a += _
  else:
    print(f"el {n} no es positivo, intenta de nuevo")

  return a

numero = sumar_elementos_1(4)

print(numero)
```

```python
# segunda forma de hacerlo

def sumar_elementos_2(n):
  a = 0

  if n >= 0:
    a = (n * (n + 1)) / 2     # -> usando la formula de Gauss
  else:
    print(f"el {n} no es positivo, intenta de nuevo")

  return a

numero = sumar_elementos_2(10)
print(numero)
```

```python
# tercera forma de hacerlo

def sumar_elementos_3(n):
  a = 0

  if n > 0:
    a = n + sumar_elementos_3(n - 1) #forma recursiva

  return a

numero = sumar_elementos_3(10)
print(numero)
```

# VAMOS A MOSTRAR UN POCO DE ESTADISTICA PARA QUE LO ENTIENDAN MEJOR 🙄

```python
# CODIGO PARA CONSEGUIR UNA MEDIDA PROMEDIO DE LO QUE TARDA UNA FUNCION
import time
import matplotlib.pyplot as plt

def tiempo_promedio(n = [0,0,0]):
  variable_1, variable_2, variable_3 = 0,0,0
  time_ini, time_fin = 0,0
  duracion_1, duracion_2, duracion_3 = 0,0,0

  for _ in range(1001):
    #funcion 1
    time_ini = time.perf_counter()
    variable_1 = sumar_elementos_1(100)
    time_fin = time.perf_counter()
    duracion_1 = time_fin - time_ini
    n[0] = n[0] + duracion_1
    del variable_1, duracion_1

    #funcion 2
    time_ini = time.perf_counter()
    variable_2 = sumar_elementos_1(100)
    time_fin = time.perf_counter()
    duracion_2 = time_fin - time_ini
    n[1] = n[1] + duracion_2
    del variable_2, duracion_2

    #funcion 3
    time_ini = time.perf_counter()
    variable_3 = sumar_elementos_1(100)
    time_fin = time.perf_counter()
    duracion_3 = time_fin - time_ini
    n[2] = n[2] + duracion_3
    del variable_3, duracion_3

  n[0], n[1], n[2] = n[0] / 1000, n[1] / 1000, n[2] / 1000

  return n

numero = tiempo_promedio()
categorias = ['Ciclo for', 'Formula de Gauss', 'Funcion recursiva']

# Crear el gráfico de barras
plt.bar(categorias, numero)

# Personalizar el gráfico
plt.xlabel('Funciones')
plt.ylabel('Tiempo (seg)')
plt.title('comparativa promedio en 1000 veces de ejecucion')

# Mostrar el gráfico
plt.show()
```

# FUNCION FILTRAR

```python
"""
La función filter() crea un iterador con los elementos de la lista que cumplen una condición.
Para filtrar los precios superiores a 100, usamos una función lambda.
"""
precios = [85, 120, 150, 95, 200, 75]

precios_altos = list(filter(lambda p: p > 100, precios))
print(f"Precios altos: {precios_altos}")
```

# FUNCION MAPEAR


```python
"""
La función map() aplica una función a cada elemento de un iterable
"""

precios_con_descuento = list(map(lambda p: p * 0.90, precios_altos))
print(f"Precios con descuento: {precios_con_descuento}")
```

# FUNCION REDUCIR

```python
"""
La función reduce() aplica una función de forma acumulativa a los elementos de un iterable.
De izquierda a derecha, para reducirlo a un solo valor.
Para usarla, necesitas importarla del módulo functools.
En este caso, sumaremos los precios con descuento para obtener el total.
"""

from functools import reduce


precio_total = reduce(lambda acumulador, p: acumulador + p, precios_con_descuento)
print(f"Precio total final: {precio_total}")
```

```python
"""
El uso de filter, map y reduce en cadena (pipeline) te permite escribir código más claro y funciona.
Procesando datos en una secuencia de transformaciones bien definidas.
"""
```

# CONCEPTO NUEVO: CLOUSURE

```python
"""
es una técnica que permite a una función anidada recordar y acceder a las variables de la función exterior.
incluso después de que la función exterior ha terminado de ejecutarse
"""
```

```python
### REGLA DE ALCANCE DE PYTHON

"""
Si una variable se define dentro de una función, se convierte en una variable local.
Si se define fuera de una función, se convierte en una variable global a nivel de módulo
"""


variable_global = "Soy una variable global."
print("hola de nuevo")

def funcion_exterior():
    # Esta es una variable local para 'funcion_exterior'
    variable_local_exterior = "Soy una variable local en la función exterior."

    def funcion_anidada():
        # Esta variable es local para 'funcion_anidada'
        variable_local_anidada = "Soy una variable local en la función anidada."

        print(f"Desde la función anidada:")
        print(f"- Puedo acceder a la variable local de mi propia función: '{variable_local_anidada}'")
        print(f"- También puedo acceder a la variable local de la función que me contiene: '{variable_local_exterior}'")
        print(f"- Y también puedo acceder a la variable global: '{variable_global}'")

    # Llamamos a la función anidada desde la función exterior
    funcion_anidada()

# Llamamos a la función exterior
funcion_exterior()
print("hola")
```

# USAREMOS ALGO DE ALCANCE GLOBAL 😂

```python
# Palabra reservada global
"""
Para modificar una variable global dentro de una función en Python, debes usar la palabra clave global.
"""


# Definimos una variable global
contador = 0

def incrementar_sin_global():
    # Esto crea una nueva variable local llamada 'contador'
    # y no afecta a la variable global
    contador = 1
    print(f"Dentro de la función (sin global): {contador}")

def incrementar_con_global():
    # Usamos la palabra clave 'global' para indicar que
    # queremos modificar la variable global 'contador'
    global contador
    contador += 1
    print(f"Dentro de la función (con global): {contador}")

# Mostrar el valor inicial
print(f"Valor inicial de la variable global: {contador}")

# Llamar a la función sin 'global'
incrementar_sin_global()
print(f"Valor después de llamar a la función sin global: {contador}") # El valor sigue siendo 0

# Llamar a la función con 'global'
incrementar_con_global()
print(f"Valor después de llamar a la función con global: {contador}") # El valor ahora es 1
```

# AHORA, ALGO NO LOCAL

```python
"""
La palabra clave nonlocal en Python se usa para indicar que una variable dentro de una función anidada no es local a esa función
y debe referirse a la variable de la función inmediatamente exterior que la contiene.
(no en un ambito global)
"""

def contador_sin_nonlocal():
    # Variable de la función exterior
    valor = 0
    def incrementar():
        # Esto crea una nueva variable LOCAL
        valor = 1
        print(f"Función anidada (sin nonlocal): {valor}")
    incrementar()
    print(f"Función exterior (sin nonlocal): {valor}")

def contador_con_nonlocal():
    # Variable de la función exterior
    valor = 0
    def incrementar():
        # Usamos nonlocal para referirnos a la variable de la función exterior
        nonlocal valor
        valor += 1
        print(f"Función anidada (con nonlocal): {valor}")
    incrementar()
    print(f"Función exterior (con nonlocal): {valor}")

print("--- Ejecución sin 'nonlocal' ---")
print(contador_sin_nonlocal())

print("\n--- Ejecución con 'nonlocal' ---")
print(contador_con_nonlocal())
```

```python
# 1. Definir la clase
class Coche:
    # El método __init__ se llama cuando se crea un nuevo objeto.
    # self se refiere al objeto que se está creando.
    def __init__(self, color, modelo):
        self.color = color
        self.modelo = modelo
        self.velocidad = 0

    # 2. Definir un método
    def acelerar(self, incremento):
        self.velocidad += incremento
        print(f"El {self.modelo} de color {self.color} va a {self.velocidad} km/h.")

# 3. Crear objetos (instancias) de la clase Coche
mi_coche = Coche("Rojo", "Ford Mustang")
tu_coche = Coche("Azul", "Tesla Model 3")

# 4. Acceder a atributos y llamar a métodos
print(f"Mi coche es un {mi_coche.modelo} de color {mi_coche.color}.")
mi_coche.acelerar(50)

print(f"Tu coche es un {tu_coche.modelo} de color {tu_coche.color}.")
tu_coche.acelerar(30)
```

```python
### metodos especiales

# __init__ : Constructor para inicializar el objeto
# __str__ : Representación legible del objeto para el usuario
# __repr__ : Representación formal y sin ambigüedades del objeto para el desarrollador
# __del__ : Destructor del objeto
# __len__ : Retorna la longitud del objeto (para la función len())
# __add__ : Sobrecarga el operador de suma (+)
# __sub__ : Sobrecarga el operador de resta (-)
# __mul__ : Sobrecarga el operador de multiplicación (*)
# __truediv__ : Sobrecarga el operador de división (/)
# __eq__ : Sobrecarga el operador de igualdad (==)
# __getitem__ : Permite indexar (objeto[índice])
# __setitem__ : Permite asignar un valor a un índice (objeto[índice] = valor)
# __iter__ : Permite iterar sobre el objeto
# __next__ : Permite obtener el siguiente elemento en una iteración
# __call__ : Permite que un objeto sea invocado como si fuera una función
```
