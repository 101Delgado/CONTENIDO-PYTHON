BACKTRACKING


```python
"""
 Imagina que est치s en un laberinto y tienes que encontrar la salida.
 Empiezas a caminar por un camino, pero si llegas a un callej칩n sin salida,
 쯤u칠 haces? Te das la vuelta, retrocedes hasta el 칰ltimo punto donde
 ten칤as una opci칩n, y tomas un camino diferente. 춰Eso, en esencia,
 es el backtracking! Es una estrategia de fuerza bruta inteligente que
 construye soluciones paso a paso, y si en alg칰n momento se da cuenta
 de que el camino actual no es viable,
 deshace los pasos (o 'backtracks') y prueba otra ruta.
 """
```

```python
def encontrar_combinaciones(numeros, objetivo):
    resultados = []

    def backtrack(combinacion_actual, suma_actual, inicio_indice):
        # Caso base 1: encontramos una soluci칩n
        if suma_actual == objetivo:
            # Agregamos una copia de la combinaci칩n actual a los resultados
            resultados.append(list(combinacion_actual))
            return

        # Caso base 2: la suma excede el objetivo, este camino no sirve
        if suma_actual > objetivo:
            return

        # Explora: recorrer los n칰meros disponibles
        for i in range(inicio_indice, len(numeros)):
            num = numeros[i]

            # Elige: agregamos el n칰mero a la combinaci칩n actual
            combinacion_actual.append(num)
            suma_actual += num

            # Explora: llamada recursiva para el siguiente n칰mero
            # Usamos `i` como nuevo inicio para evitar duplicados
            print(combinacion_actual)
            backtrack(combinacion_actual, suma_actual, i)

            # Deshace: si el camino no funcion칩, quitamos el 칰ltimo n칰mero
            combinacion_actual.pop()
            suma_actual -= num

    # Iniciar el proceso de backtracking
    backtrack([], 0, 0)

    return resultados

# Ejemplo de uso
numeros = [2, 3, 6, 7]
objetivo = 8
print(f"Combinaciones para la suma {objetivo}: {encontrar_combinaciones(numeros, objetivo)}")
# Salida esperada: [[7], [2, 2, 3]]
```

```python
def min_monedas(monedas, monto):
    # dp[i] almacenar치 el n칰mero m칤nimo de monedas para el monto 'i'
    dp = [float('inf')] * (monto + 1)
    dp[0] = 0 # Se necesitan 0 monedas para hacer un monto de 0

    print(dp)

    # Llenamos la tabla de forma bottom-up
    for i in range(1, monto + 1):
        # Para cada monto 'i', probamos cada moneda
        for moneda in monedas:
            # Si el monto actual es mayor o igual a la moneda...
            if i - moneda >= 0:
                # ...actualizamos el valor de dp[i] si encontramos una soluci칩n mejor
                # La soluci칩n es 1 (la moneda actual) + el m칤nimo de monedas para el resto del monto
                dp[i] = min(dp[i], 1 + dp[i - moneda])

    print(dp)

    # Si dp[monto] sigue siendo infinito, significa que no se puede formar el monto
    return dp[monto] if dp[monto] != float('inf') else -1

# Ejemplo de uso
monedas = [1, 2, 5]
monto = 11
print(f"M칤nimo de monedas para el monto {monto}: {min_monedas(monedas, monto)}")
# Salida esperada: 3 (5 + 5 + 1)
```
