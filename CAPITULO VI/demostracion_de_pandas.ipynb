```python
import numpy as np

data = {
    "estudiantes" : ["juan", "johan", "gabriel", "jesus"],
    "ingresos anuales" : [1000, 2000, np.nan, 20],
    "id" : [1, np.nan, 2, 3]
}

df = pd.DataFrame(data)

print(df)
print(df.isnull().sum())

df['ingresos anuales'].fillna(df['ingresos anuales'].mean(), inplace= True)

print(df)
print(df['ingresos anuales'].mean())

df['id'].fillna("desconocido", inplace = True)
print(df)
```
