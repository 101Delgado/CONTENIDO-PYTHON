```python
"""
EJEMPLO
"""
"""
ALGORITMO DE CAMBIO DE MONEDAS
Usando la menor cantidad de mondedas posibles, indique
la totalidad y cuales monedas se utilizaron
"""
```

```python
coins = [500, 100, 50, 20, 10, 5, 1]

def change(cantidad): #funcion principal
  #-------- variables ---------
  lista = []  #lista que saldra con la monedas que se utilizaron
  indice = 0  # indice que nos ayudara a movernos por la lista

  #-------- ciclo prinpipal---
  while cantidad > 0: #reducieremos la cantidad conforme le restemos las monedas
    if cantidad < coins[indice]:  # si la moneda es mayor que la cantidad
      #avanzamos a la siguiente moneda
      indice += 1
    else: # si la cantidad es mayor que la moneda
      # agregamos la moneda a la list
      lista.append(coins[indice])
      # le restamos la denominacion de la moneda a la cantidad
      cantidad -= coins[indice]

  return lista

# algoritmo principal
cantidad = 652
cambio = change(cantidad)
print(f"todas las monedas que utilizamos son {cambio}")
```

```python
"""
problemas del codigo:
"""
# si la lista de las monedas esta desordenada pierde efectividad
# solo determina la mejor opcion para el momento
# tendremos que hacer un paso extra para ordenar la lista de mayor a menor
```

```python
import random as rd

rd.shuffle(coins)
cantidad = 652
cambio = change(cantidad)
print(cambio)
```

ENFOQUE DIVIDE Y CONQUISTA

```python
"""
El enfoque "Divide y Conquista" (Divide and Conquer) es una estrategia
algortmica poderosa que resuelve un problema complejo dividindolo
en subproblemas ms pequeos y manejables.
"""
# YA NOSOTROS HEMOS HABLADO DE ESO EN VARIAS DE NUESTRAS CLASES
```

```python
"""
Dividir (Divide): El problema original se descompone en varios subproblemas
ms pequeos del mismo tipo. Estos subproblemas son ms sencillos
de resolver que el problema original.

Conquistar (Conquer): Se resuelven los subproblemas de forma recursiva.
Si un subproblema es lo suficientemente pequeo, se resuelve directamente; si no,
se divide de nuevo hasta que se pueda resolver.

Combinar (Combine): Las soluciones de los subproblemas se unen
para construir la solucin del problema original.
"""
```

```python
# podemos hacerlo recursivo hasta que nos quede una matriz 2x2

def Llenado(tablero_aux, relleno = 2):

  if (len(tablero_aux[0]) > 1):
    bloques_verticales = np.vsplit(tablero_aux, 2)

    # vsplit nos devuelve una lista con las dos matrices
    bloque_arriba = bloques_verticales[0]
    bloque_abajo = bloques_verticales[1]

    # 3. Dividir horizontalmente cada bloque para obtener los cuatro bloques 4x4
    bloque_arriba_izq, bloque_arriba_der = np.hsplit(bloque_arriba, 2)
    bloque_abajo_izq, bloque_abajo_der = np.hsplit(bloque_abajo, 2)

    # Guardar los cuatro bloques en una lista para fcil acceso
    bloques = [bloque_arriba_izq, bloque_arriba_der,
              bloque_abajo_izq, bloque_abajo_der]

    if (bloque_arriba_izq == 0).all():
      if (bloque_abajo_izq == 0).all():
        if (bloque_abajo_der == 0).all():
          bloque_arriba_izq[len(bloque_arriba_izq) - 1][len(bloque_arriba_izq) - 1] = relleno
          bloque_abajo_izq[0][len(bloque_abajo_izq) - 1] = relleno
          bloque_abajo_der[0][0] = relleno
        else:
          bloque_arriba_izq[len(bloque_arriba_izq) - 1][len(bloque_arriba_izq) - 1] = relleno
          bloque_arriba_der[len(bloque_arriba_der) - 1][0] = relleno
          bloque_abajo_izq[0][len(bloque_abajo_izq) - 1] = relleno
      else:
        bloque_arriba_izq[len(bloque_arriba_izq) - 1][len(bloque_arriba_izq) - 1] = relleno
        bloque_arriba_der[len(bloque_arriba_der) - 1][0] = relleno
        bloque_abajo_der[0][0] = relleno
    else:
      bloque_arriba_der[len(bloque_arriba_der) - 1][0] = relleno
      bloque_abajo_izq[0][len(bloque_abajo_izq) - 1] = relleno
      bloque_abajo_der[0][0] = relleno

    Llenado(bloque_arriba_izq, relleno + rd.randrange(0,100))
    print(tablero,"\n")
    Llenado(bloque_abajo_izq, rd.randrange(0,100))
    print(tablero,"\n")
    Llenado(bloque_abajo_der, relleno + rd.randrange(0,100))
    print(tablero,"\n")
    Llenado(bloque_arriba_der, relleno + rd.randrange(0,100))

Llenado(tablero)
tablero
```
