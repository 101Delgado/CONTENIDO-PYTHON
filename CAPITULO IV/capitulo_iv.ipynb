# QUE ES UNA PILA? 游뿲
En Venezuela las baterias tambien se llaman pilas

```python
"""
Es una estructura de datos abstracta que sigue el principio LIFO (Last-In, First-Out).
Lo que significa "el 칰ltimo en entrar es el primero en salir".
Es como una pila de platos: el 칰ltimo plato que pones encima es el primero que quitas.
"""

# debe cumplir con dos op칠raciones basicas

"""
Push: Agrega un nuevo elemento a la parte superior de la pila.

Pop: Elimina el elemento que est치 en la parte superior de la pila.
"""

# otras dos operaciones utiles son

"""
Peek (o Top): Permite ver el elemento superior de la pila sin eliminarlo.

IsEmpty: Verifica si la pila est치 vac칤a.
"""
```

```python
"""
una estructura de datos abstracta que sigue el principio FIFO (First-In, First-Out),
que significa "el primero en entrar es el primero en salir".
Es exactamente como una fila de personas esperando
"""


# Las operaciones de una cola se realizan en extremos opuestos, lo que la diferencia de una pila:

"""
Enqueue: Agrega un nuevo elemento al final de la cola.

Dequeue: Elimina el elemento que se encuentra al principio de la cola.
"""

# Adem치s, hay otras operaciones 칰tiles:

"""
Peek: Permite ver el elemento que est치 al frente de la cola sin eliminarlo.

IsEmpty: Comprueba si la cola est치 vac칤a.
"""
```

Declaremos entonces como son las colas

```python
class Queue():
  def __init__(self, *contenido):
    self.contenido = list(contenido)

  def enqueue(self, elemento): #a침ade un elemento al final de la cola
    self.contenido.append(elemento)

  def dequeue(self): # saca el elemento que esta al principio
    if (self.IsEmpty):
      return None
    else:
      return self.contenido.pop(0)

  def peek(self): #nos muestra el elemento que esta al principio de la cola
    if (self.contenido.IsEmpty()):
      return None
    else:
      return self.contenido[0]

  def IsEmpty(self):  #nos indica si la cola esta vacia
    if (len(self.contenido) == 0):
      return True
    else:
      return False
  #extra
  def size(self): #devuelve el tama침o de la cola
    return len(self.contenido)

  def __str__(self) -> str: #devuelve el contenido de la cola
    return str(self.contenido)
```

```python
def seq_search(estr, dato):
  for _ in estr:
    if (dato == _):
      return estr.index(_)

dato = seq_search([2,3,4,5,6],4)
print(dato)
```

# BUSQUEDA BINARIA

```python
"""
Funciona dividiendo la lista en dos mitades y comparando el valor buscado con el elemento central.
Si el valor es menor, la b칰squeda contin칰a en la mitad izquierda de la lista;
si es mayor, en la mitad derecha. Este proceso se repite hasta encontrar el elemento.
"""
```

```python
# ALGORITMO DE BUSQUEDA DE DATOS

def bin_search(estr, dato):
  bajo = 0
  alto = len(estr) - 1

  while bajo <= alto:
    medio = (bajo + alto) // 2

    if estr[medio] == dato:
      return medio
    elif estr[medio] > dato:
      alto = medio - 1
    else:
      bajo = medio + 1
  return -1

indice = bin_search([0,1,2,3,4,5,6],5)
indice
```

```python
nums = [9, 13, 18,34,55, 19, 30, 32, 34, 36, 38, 49]

def ordenamiento(lista):
  aux = []
  for _ in lista:
    if len(aux) == 0:
      aux.append(_)
    else:
      for i in aux:
        if _ > i:
          if aux.index[i] == len(aux) - 1:
            aux.append(_)
          else:
            if _ < aux[aux.index(i) + 1]:
              aux.insert(aux.index(i) + 1, _)
        else:
          pass
  return aux

  numero = ordenamiento(nums)
  print(numero)
```
