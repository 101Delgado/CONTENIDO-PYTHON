CONJUNTOS DENTRO DE PYTHON

```python
# declaracion de conjuntos
A = set(range(13))
B = set(range(0,13,3))

#interseccion de conjuntos
inter_AB = A & B
print(f"interseccion de A con B {inter_AB}")

#union de conjuntos
union_AB = A | B
print(f"union de A con B {union_AB}")

# como saber si un conjunto es subconjunto de otro

# revisemos la interseccion
# si la interseccion de A con B es el mismo conjunto B
# quiere decir que todos los elementos de B estan en A
if (inter_AB == B):
  print(f"{B} es subconjunto de {A}")
```

CALCULO DE FUNCIONES TRIGONOMETRICAS

```python
import math
# recomendacion, pasenlo todo por radianes y no por grados
angulo = 90
angulo = math.radians(angulo)

coseno = math.cos(angulo)
seno = math.sin(angulo)


print(seno)
print(coseno)
```

FUNCION SIGMOIDEA

```python
import numpy as np

def funcion_sigmoidea(x):
    """Calcula la función sigmoidea (logística) para un valor o array de valores."""
    return 1 / (1 + np.exp(-x))
```

```python
# utilizando la libreria sympy

from sympy import symbols, solve

# declaramos la variable
x = symbols('x')

# la ecuacion 3x + 5 = 1
# la igualamos a 0

ecuacion = 3*x + 4

solucion = solve(ecuacion,x)

print(solucion)
```
