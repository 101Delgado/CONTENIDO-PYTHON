```python
""" revisaremos los metodos disponibles para las listas en python"""

"""metodos para aÃ±adir elementos"""
lista = list(range(4))

print(f"lista original -> {lista}")

#.append()
"""aÃ±ande un elemento al final de una lista"""

lista.append("ssss")

print(f"lista utilzando append -> {lista}")

#.extend()
"""aÃ±ade un objeto iterable al final de la lista"""
"""podemos aÃ±adir otra lista o tuplas"""

lista.extend(range(2))

print(f"lista utilizando extend -> {lista}")

#.insert()
"""aÃ±ade un elemento en una posicion especifica de la lista"""
"""si el indice supera el que permite la lista lo inserta al final"""

lista.insert(8, "aÃ±adido")

print(f"lista utilizando insert -> {lista}")
```

```python
""" metedos para eliminar elementos"""

#.remove()
"""remueve la primera coincidencia de un valor"""

lista.remove(1)

print(f"lista utilizando remove -> {lista}")

#.pop()
"""elimina el valor en una posicion especifica y lo devuelve"""
"""si no se especifica el indice elimina el ultimo valor de la lista"""

valor = lista.pop()

print(f"lista utilizando pop -> {lista} se elimino el {valor}")

#.clear()
"""elimina todos los elementos de la lista dejandola vacia"""
lista_aux = list(range(5))

lista_aux.clear()

print(f"lista utilizando clear -> {lista_aux}")
```

```python
"""metodo de ordenacion y reversion """
lista = list(range(5))

#.sort()
"""solo ordena listas con numeros"""
lista.sort()

print(f"lista utilizando sort -> {lista}")

#.reverse()
"""ordena de forma reversa la lista"""
lista.reverse()

print(f"lista utilizando reverse -> {lista}")
```

```python
lista_auxilar = [2,0,1,5,"Jose"]
variable = lista_auxilar.pop()
lista_auxilar.sort()
lista_auxilar.append(variable)
lista_auxilar
```

```python
lista_auxilar = [2,0,1,5,"Jose"]
lista_auxilar[:4].sort()
print(lista_auxilar)

print(f"id de {id(lista_auxilar)} y id de {id(lista_auxilar[:4])}")
```

```python
lista_3 = [1,2,1 + 4j,16]
lista_3.sort()
```

```python
"""metodos de consula"""

#.count()
"""cuenta la cantidad de veces que aparece un determinado elemento"""

cont = lista.count(1)

print(f"lista utilizando count -> {lista} el 1 aparecio {cont} veces")

#.copy()
"""copia el contenido de la lista"""
"""no modifica la lista original"""

lista_nueva = lista.copy()

print(f"lista utilizando copy -> {lista} y la otra lista es {lista_nueva}")

#.index()
"""retorna el indice de la primera aparicion de un elemento"""

indice = lista.index(1)

print(f"lista utilizando index -> {lista} el 1 tiene indice {indice}")
```

# **DICCIONARIOS DENTRO DE PYTHON**

```python
diccionario = {2 : [5,5,5] , "carne" : ["vaca" , "pollo"] , "4": "d" , "marcelo" : 2}

""" se guardan mediante un par clave valor"""
""" no admite claves duplicadas, sobreescribe la utlimna"""

#acedemos a los datos por su clave

print(diccionario[2])

#podemos sobreescribirlo mediante la clave

diccionario[2] = "america"

print(diccionario[2])

#si el indice no existe se crea un nuevo apartado

diccionario[3] = "Danny"

print(diccionario)
```

```python
"""revisaremos los metodos de los diccionarios dentro de python"""

"""metodos para aÃ±adir y actualizar"""

#.update()
"""inserta un diccionario dentro de otro """
diccionario1 = {2 : 5}
diccionario.update(diccionario1)

print(f"usando update -> {diccionario}")

#.fromkeys()
"""crea un diccionario nuevo con una lista de claves y con un valor dado"""
claves = [5,4,6,8]
diccionario1 = dict.fromkeys(claves, 0)

print(f"usando fromkeys -> {diccionario1}")
```

```python
"""metodos de eliminacion"""
#.pop()
"""Elimina la clave especificada y devuelve su valor"""

diccionario.pop(2)
print(f"usando pop -> {diccionario}")

#.popitem()
"""elimina y devuelve el ultimo par clave- valor ingresado en el diccionario"""

clave_valor = diccionario.popitem()

print(f"usando popitem -> {diccionario}, nos quedo el par {valor}")

#.clear()
"""limpia el diccionario """
diccionario1 = {2 : "s"}

diccionario1.clear()

print(f"usando clear -> {diccionario1}")
```

```python
"""metodos de acceso y vista """

#.get()
"""devuelve el valor para una clave dada"""

valor = diccionario.get(4)

print(f"usando get -> {valor}")

#.keys()
"""devuelve un objeto vista con todas las claves"""

valor = diccionario.keys()

print(f"usando keys -> {valor}")

#.values()
"""devuelve un objeto vista con todos los valores"""

valor = diccionario.values()

print(f"usando values -> {valor}")

#.items()
"""devuelve un objeto vista con todos los pares clave valor que tiene el diccionario"""

valor = diccionario.items()

print(f"usando items -> {valor}")
```
